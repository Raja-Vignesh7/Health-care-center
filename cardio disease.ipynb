```python
import pandas as pd
import numpy as np
cols = ['Unnamed: 0', 'id', 'age', 'gender', 'height', 'weight', 'ap_hi','ap_lo', 'cholesterol', 'gluc', 'smoke', 'alco', 'active', 'cardio']
df = pd.read_csv("health_data.csv",header=None,skiprows=1,names=cols)
df
```

```python
df2 = df.drop(columns=['Unnamed: 0','id'])
df2.head()
```

```python
df2['age'] = (df['age']/365.25).astype(int)
df2.head()
```

```python
df2.describe()
```

```python
df2['BMI'] = (df2['weight']/(df2['height']*df2['height']))*10000
df2.head()
```

# BMI Range
- below 18.5 underweight
- 18.5-24.9 healthy weight
- 25-29.9 overweight
- above30 obesity

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.hist(df2['BMI'],bins=50,rwidth=0.85,log=True)
plt.title("BMI Range")
plt.xlabel("BMI value")
plt.ylabel("frequency")
plt.show()
```

```python
%matplotlib inline
df3 = df2[(df2['BMI']>10) & (df2['BMI']<50)]
plt.hist(df3['BMI'],bins=50,rwidth=0.85)
plt.show()
```

```python
%matplotlib inline
import seaborn as sns
correlation_matrix = df3.corr()
sns.heatmap(correlation_matrix,annot=True,cmap='coolwarm',fmt='.2f')
plt.show()
```

```python
%matplotlib inline
plt.hist(df3['ap_lo'],bins=50,rwidth=0.85,color="blue",log=True,label="ap_lo")
plt.hist(df3['ap_hi'],bins=50,rwidth=0.85,color="red",log=True,label="ap_hi")
plt.legend()
plt.title("ap_lo & ap_hi")
plt.xlabel("value")
plt.ylabel("freqency")
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
score_RF = cross_val_score(RandomForestClassifier(),df7.drop(columns=['cardio']),df7['cardio'])
score_RF
```

```python
from sklearn.model_selection import train_test_split
X = df7.drop(columns=['cardio'])
Y = df7['cardio']
X_train,X_test,y_train,y_test = train_test_split(X,Y,test_size = 0.3,random_state=42)
rf_model = RandomForestClassifier()
rf_model.fit(X_train,y_train)
y_pred = rf_model.predict(X_test)
y_pred
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
np.unique(df7,axis=0,return_counts=True)
```

```python
from sklearn.calibration import CalibratedClassifierCV
rf = RandomForestClassifier(n_estimators=100, random_state=42)
classifier_rf = CalibratedClassifierCV(estimator=rf , method='sigmoid' , cv = 5)
classifier_rf.fit(X_train,y_train)
prob_pos = classifier_rf.predict_proba(X_test)
prob_pos
```

```python
prob_pos.shape
```

```python
y_test
```

```python
y_test.shape
```

```python
from sklearn.metrics import brier_score_loss , accuracy_score
brier_score = brier_score_loss(y_test == 1,prob_pos[:, 1])
accuracy = accuracy_score(y_test, classifier_rf.predict(X_test))
print(f'Brier Score: {brier_score}')
print(f'Accuracy: {accuracy}')
```
